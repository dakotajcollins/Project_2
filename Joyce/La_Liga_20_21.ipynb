```python
import pandas as pd
```

```python
# --Target table url
url = 'https://fbref.com/en/comps/12/10731/2020-2021-La-Liga-Stats'
```

```python
tables = pd.read_html(url)
league_table_df = tables[0]
league_table_df.head()
```

```python
# Clean the Table
columnsplit=league_table_df['Top Team Scorer'].str.split("-", expand=True)
league_table_df=league_table_df.assign(Top_Scorer=columnsplit[0],Score=columnsplit[1])
new_league_table_df=league_table_df.drop(['Top Team Scorer','Notes'] , axis=1)
new_league_table_df.head()
```

```python
#Pick the columns we needed and assign into new datafrmae
new_league_table_df=new_league_table_df[['Rk','Squad','MP','W','D','L','GF','GA','GD',
                                    'Pts']]
new_league_table_df.head()
```

```python
#  Import other cvs file
csv_file = "sportsref_20_21.csv"
player_standard_stats_df = pd.read_csv(csv_file, header=1, encoding='maccentraleurope')
player_standard_stats_df.head()
```

```python
# Clean dataframe, split "Nation" column and only get last 3 letter
columnsplit=player_standard_stats_df['Nation'].str.split(" ", expand=True)
player_standard_stats_df=player_standard_stats_df.assign(Nation=columnsplit[1])
player_standard_stats_df.head()
```

```python
#Pick the columns we needed and assign into new datafrmae
new_player_df=player_standard_stats_df[['Player','Nation','Pos','Squad','Age','Born','MP','Starts',
                                       'Min','Gls','Ast','PK','CrdY','CrdR']]
new_player_df.head()
```

```python
# Repleace the "Min" column from int to str
new_player_df['Min']=new_player_df['Min'].str.replace(',','')
new_player_df.head()
```

```python
# Connect to local database

from sqlalchemy import create_engine
rds_connection_string = "postgres:postgres@localhost:5432/la_liga_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
#Check for tables
engine.table_names()
```

```python
# Use pandas to load csv into database
new_league_table_df.to_sql(name='league_table', con=engine, if_exists='append', index=False)
new_player_df.to_sql(name='player_standard_stats', con=engine, if_exists='append', index=False)
```

```python
# #querying  Join the two table on Squad col
Join_df=pd.read_sql_query('select * from league_table as lt join player_standard_stats as ps on lt."Squad"=ps."Squad"', con=engine)
Join_df.head()
```

```python
# Pick the column we needed from the join datafrmae
Top_player=Join_df[['Squad','Player','Age','Gls','Ast','CrdY','CrdR']]
Top_player=Top_player.iloc[:,1:]
Top_player.head()
```

```python
#groupby the "Squad" col and find the top score player
df2 =Top_player.groupby(['Squad','Player'])['Gls'].agg(['max'])
df2.head(10)
```

```python
#Sort_values by "max" col
df2 = df2.sort_values('max', ascending=False)
df2.head(20)
```

```python
#resrt_index and remove the duplicates
reset_df=df2.reset_index()
reset_df.drop_duplicates(['Squad'], keep='first')
reset_df.head(20)
```

```python
# Rename the "max" column 
TGS_df=reset_df.rename(columns={'max':'Goals Scored'})
TGS_df.head(20)
```

```python
#insert new datafrmae into database new table 
TGS_df.to_sql(name='Top_Goal_Scorer', con=engine, if_exists='append', index=False)
```
